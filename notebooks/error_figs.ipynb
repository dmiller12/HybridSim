```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../data/res.csv")
```

```python
tracked_dims = ["x", "z"]
fig, ax = plt.subplots(1, len(tracked_dims), constrained_layout=True, figsize=(8, 3))
for i, cart_i in enumerate(tracked_dims):
    ax[i].plot(data['time'], data[f"{cart_i}_d"], label="desired")
    ax[i].plot(data['time'], data[cart_i], label="actual")
    ax[i].set_ylabel(cart_i)
    ax[i].set_xlabel("time (s)")
    ax[i].legend()

fig.suptitle("Motion tracking")
```

```python
tracked_dims = ['x', 'z']
fig, ax = plt.subplots(1, len(tracked_dims), constrained_layout=True, figsize=(8, 3))
for i, cart_i in enumerate(tracked_dims):
    ax[i].plot(data['time'], data[f"f_{cart_i}_d"], label="desired")
    ax[i].plot(data['time'], data[f"f_{cart_i}"], label="actual")
    ax[i].set_ylabel(cart_i)
    ax[i].set_xlabel("time (s)")
    ax[i].legend()

fig.suptitle("Force tracking")
```
